# Firing Rate Development

```python
exc_rates_np, inh_rates_np = helper.nested_list_of_epochs_2_np(rates_subfolders[0])
exc_rates_obj = combine.average_responses(exc_rates_np, list(stimulus_ids.values()), axis=1)
inh_rates_obj = combine.average_responses(inh_rates_np, list(stimulus_ids.values()), axis=1)
```

```python
#splits by number of stimuli (so with 40 it looks mental). Basically a measure of sparsity; which neurons have a firing rate other than zero
minimal_rate_to_be_on = 10
exc_on = np.count_nonzero((exc_rates_obj > minimal_rate_to_be_on), axis=3)
inh_on = np.count_nonzero((inh_rates_obj > minimal_rate_to_be_on), axis=3)

spikeplot.plot_development_for_object(exc_on, "Number of active excitatatory neurons", list(stimulus_ids.keys()))
spikeplot.plot_development_for_object(inh_on, "Number of active inhibitory neurons", list(stimulus_ids.keys()))
```

```python
exc_mean = np.mean(exc_rates_obj, axis=-1)
inh_mean = np.mean(inh_rates_obj, axis=-1)

spikeplot.plot_development_for_object(exc_mean, "Mean fr of excitatatory neurons", list(stimulus_ids.keys()))
```
